```python
movies['movie_duration_category'] = pd.cut(
    x=movies['duration'],
    bins=[1, 90, 120, np.inf],
    labels=['1. <90','2. 90–120','3. >120'])

movies['movie_duration_category'].head() #Task 1.5 - create movie_duration_category column, which will contain three categories
```

```python
movies['plot_keywords']
```

```python
movies['plot_keywords'].fillna('other', inplace=True)
```

```python
movies['plot_keywords'].sample(20)
```

```python
keywords = ['love','death']
movies['movie_plot_category'] = movies['plot_keywords'].str.findall('|'.join(keywords)).apply(set).str.join(' and ')

movies['movie_plot_category'].sample(20) #Task 1.8 - create 'love, death, other' movie plot cateogry column 
```

```python
movies.movie_plot_category.replace('','other',regex = True)
movies['movie_plot_category'].sample(20)
```

```python
np.round(movies.groupby('movie_plot_category')['imdb_score'].mean(),2) #Task 1.9 & 1.10, the average imdb score
```

```python
movies['budget'] = movies['budget'].str.replace('$', '')
movies['budget'].sample(20)
```

```python
movies['budget'] = movies['budget'].astype(float)
movies['budget'] = movies['budget'].astype(int)
movies['budget'].sample(20)
```
